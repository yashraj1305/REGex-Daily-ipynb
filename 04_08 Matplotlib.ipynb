## Bar Graph 

```python
plt.figure()
x = np.arange(10)
y = x*2
plt.bar(x,y, width=0.8)
plt.xticks(x)
plt.yticks(y)
```

```python
plt.figure()
x = np.arange(10)
y = x*2
z = x*4
plt.bar(x,y, width=0.4 , color = 'm')
plt.bar(x,z, width=0.4 , color = 'c' , bottom=x)
```

```python
plt.figure()
x = np.arange(10)
y = x*2
z = x*4
plt.bar(x-0.2,y, width=0.4 , color = 'm')
plt.bar(x+0.2,z, width=0.4 , color = 'c')
```
